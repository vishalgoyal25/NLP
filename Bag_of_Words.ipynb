```python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection',
                    sep='\t',names=["label","message"])
```

```python
messages
```

## Data Cleaning And Preprocessing

```python

import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
corpus=[]

for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ', messages['message'][i])          ## Apart from a-z and A-Z, Replace everything with Blank.

    review=review.lower()
    review=review.split()               ## This Creates a List of Reviews[]

    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    
    review=' '.join(review)

    corpus.append(review)
```

```python
corpus
```

## Create a Bag Of Words (BoW)

```python
## Create the Bag OF Words model
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
```

```python
## for Binary BOW enable binary=True

cv=CountVectorizer(max_features=100,binary=True)
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
cv.vocabulary_
```
